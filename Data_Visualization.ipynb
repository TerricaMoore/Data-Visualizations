```
import numpy as np #Numpy provides fast mathematical function processing
import pandas as pd #Pandas is used for eacy indexing and data collection tools
import matplotlib as mpl # High level data Visualization library
import matplotlib.pyplot as plt #High level data Visualization library
%matplotlib inline
```

```
#My data is called Loans so I named "Loans" my dataframe (Change the name of yours relative to your data)
#the below is the code to import csv in pythong
Loans=pd.read_csv("https://raw.githubusercontent.com/TerricaMoore/Data-Visualizations/main/Loan%20payments%20data.csv",encoding='latin1')
```

```
Loans2 = Loans.plot.line(x='terms', y='Principal')

plt.title('Line Plot')
plt.xlabel('Terms')
plt.ylabel('Principal')
plt.show
```

```
Loans3 = Loans.plot.bar(x='terms', y='Principal')

plt.title('Bar')
plt.xlabel('Terms')
plt.ylabel('Principal')
plt.show
```

```
Loans4 = Loans.plot.barh(x='terms', y='Principal')

plt.title('Bar')
plt.xlabel('Terms')
plt.ylabel('Principal')
plt.show
```

```
Loans11 = Loans.plot.hist(x='Gender', y='Principal')

plt.title('Bar')
plt.xlabel('Gender')
plt.ylabel('Principal')
plt.show
```

```
Loans5 = Loans.plot.hist(x='Gender', y='Principal')

plt.title('Bar')
plt.xlabel('Gender')
plt.ylabel('Principal')
plt.show
```
